## Investigation Overview

>In this presentation, we will present some analyzes and visualizations of the data features to see how they are related to the borrower's annual percentage rate
We'll show some visualizations of features like BorrowerAPR, Prosper Score, and Credit Score Range Upper.

## Dataset Overview

>We have a data set containing 113,937 loans with 81 variables for each loan This includes loan amount, interest rate, current loan status, borrower income, and many other variables.

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
df=pd.read_csv('ProsperLoanData.csv')
df.head()
```

## Borrowers Rating vs Income Range
The violin chart shows that there is no strong relationship between the current income level of the borrower and the interest rate, as the average relationship ranges between 15 to 25% according to the current income rate.

```python
plt.figure(figsize = [10, 7])
plt.subplots_adjust(top = 3.5, bottom = 2);

plt.subplot(2,1,2)
sb.violinplot(data = df, y = 'IncomeRange', x = 'BorrowerRate', palette = "flare",  inner='quartile')
plt.title("Current Rate vs. Income Classes");


```

## BorrowerAPR vs. ProsperRating
> For Prosper Score, there is clearly a negative association with BorrowerAPR
But there is not much overlap in the case of ProsperRating for the two variables

```python
plt.figure(figsize = [20,8]) 
colorbase= sb.color_palette()[2]
plt.gcf().set_size_inches(10, 8)


plt.subplot(1, 2, 2)
sb.boxplot(data = df, x = 'ProsperScore', y = 'BorrowerAPR',color=colorbase)
plt.title('BorrowerAPR vs ProsperScore')
plt.xlabel('ProsperScore')
plt.ylabel('BorrowerAPR');

plt.subplot(1, 2, 1)
sb.boxplot(data = df, x = 'ProsperRating (Alpha)', y = 'BorrowerAPR',color=colorbase)
plt.title('BorrowerAPR vs. ProsperRating')
plt.xlabel('ProsperRating')
plt.ylabel('BorrowerAPR')
```

## BorrowerAPR & CreditScoreRangeUpper & ProsperScore
> The effect of borrowerAPR . can be observed
We can see the creditScoreRangeUpper rise with the borrowerAPR fall in parcels
By adding ProsperScore, BorrowerAPR is reduced and vice versa
This proves that CreditScoreRangeUpper and ProsperScore are adversely correlated with BorrowerAPR.


```python
plt.figure(figsize = [15, 9]) 
sb.stripplot(data = df, x = 'CreditScoreRangeUpper', y = 'BorrowerAPR', 
             hue ='ProsperScore', jitter = 0.35 , dodge = True , palette ="Set2")

plt.title('BorrowerAPR & CreditScoreRangeUpper & ProsperScore')
plt.xlabel('CreditScoreRangeUpper')
plt.ylabel('BorrowerAPR')
```
